```python
from loguru import logger
import ollama
class OllamaQW():
    tools = []
    base_message = [
            {"role": "system", "content": "你是Alice,是YakumoAki在设计的智能语音助手"}
        ]
    history_message = []
    messages = []
    def __init__(self, model_name: str):
        import yaml
        self.model_name = model_name
        # 打开并读取YAML文件
        with open('ToolConfig.yml', 'r', encoding='utf-8') as file:
            self.tools = yaml.load(file, Loader=yaml.FullLoader)
        logger.info(config_yml)
        
    def add_message(self, role,message):
        self.history_message.append({
            "role": role,
            "content": message,
        })
        self.messages =self. base_message + self.history_message[:-20]
    def add_message(self, message):
        self.history_message.append(message)
        self.messages =self. base_message + self.history_message[:-20]
        
    def get_tool_response(self,response):
        self.messages.append(response["message"])
        if tool_calls := self.messages[-1].get("tool_calls", None):
            for tool_call in tool_calls:
                if fn_call := tool_call.get("function"):
                    fn_name: str = fn_call["name"]
                    fn_args: dict = fn_call["arguments"]

                    fn_res: str = globals()(fn_name)(**fn_args)

                    self.messages.append({
                        "role": "tool",
                        "name": fn_name,
                        "content": fn_res,
                    })
        response = ollama.chat(
            model=self.model_name,
            messages=self.messages,
            tools=self.tools,
        )
        self.messages.append(response["message"])
    
    
    
    def chat(self, msg):
        self.add_message("user", msg)
        response = ollama.chat(
            model=self.model_name,
            messages=self.messages,
            tools=self.tools,
        )
        self.add_message(response["message"])
        if tool_calls := self.messages[-1].get("tool_calls", None):
            for tool_call in tool_calls:
                if fn_call := tool_call.get("function"):
                    fn_name: str = fn_call["name"]
                    fn_args: dict = fn_call["arguments"]

                    fn_res: str = globals()(fn_name)(**fn_args)

                    self.add_message({
                        "role": "tool",
                        "name": fn_name,
                        "content": fn_res,
                    })
        response = ollama.chat(
            model=self.model_name,
            messages=self.messages,
            tools=self.tools,
        )
        self.add_message(response["message"])
        return self.messages[:-1]
```

```python
def api_generate(text:str):
  print(f'提问：{text}')
 
  stream = ollama.generate(
    stream=True,
    model='qwen2.5:1.5b',
    prompt=text,
    )
 
  print('-----------------------------------------')
  for chunk in stream:
    if not chunk['done']:
      print(chunk['response'], end='', flush=True)
    else:
      print('\n')
      print('-----------------------------------------')
      print(f'总耗时：{chunk['total_duration']}')
      print('-----------------------------------------')

# 流式输出
api_generate(text='天空为什么是蓝色的？')
```
